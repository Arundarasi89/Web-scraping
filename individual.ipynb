```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
url='https://insights.blackcoffer.com/do-all-social-media-is-owned-by-meta/'
```

```python
headers = {'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36"}
```

```python
requests.get(url,headers=headers)
```

```python
p = requests.get(url,headers=headers)
```

```python
soup = bs4.BeautifulSoup(p.text, 'html.parser')
```

```python
soup.title
```

```python
title = soup.title.string
print(title)
```

```python
titles = soup.find("h1", class_="entry-title").getText()
print(titles)
```

```python
article_body = soup.find("div", class_="td-post-content").getText()
print(article_body)
```

```python
c = titles + article_body
print(c)
```

```python
with open('text_files/1.txt', 'w') as f:
  f.write(c)
  f.write('\n')
  f.close()
```
