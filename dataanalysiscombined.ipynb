```python
#List of stopwords
a = open('allstopwords.txt', 'r')
b = a.readlines()

new_stopwords = []
for line in b:
    for item in line.strip().split():
        new_stopwords.append(item)
print(new_stopwords)
```

```python
#Tokenization and removal of stopwords
def tokenizer(data):
    text = data
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)
    filtered_words = list(filter(lambda token: token not in new_stopwords, tokens))
    return filtered_words
```

```python
#Loading positive words
with open('pw.txt','r') as posfile:
    positivewords=posfile.read()
positiveWordList=positivewords.split('\n')
```

```python
#Loading negative words
with open('nw.txt' ,'r',encoding="ISO-8859-1") as negfile:
    negativeword=negfile.read()
negativeWordList=negativeword.split('\n')
```

```python
#Calculating positive score 
def positive_score(text):
    numPosWords = 0
    rawToken = text
    for word in rawToken:
        if word in positiveWordList:
            numPosWords  += 1
    
    sumPos = numPosWords
    return sumPos
```

```python
#Calculating negative score 
def negative_word(text):
   # str_tok = ' '.join([str(item) for item in text])
    numNegWords=0
    rawToken = text
    for word in rawToken:
        if word in negativeWordList:
            numNegWords -=1
    sumNeg = numNegWords 
    sumNeg = sumNeg * -1
    return sumNeg
```

```python
# Calculating polarity score
def polarity_score(positiveScore, negativeScore):
    pol_score = (positiveScore - negativeScore) / ((positiveScore + negativeScore) + 0.000001)
    return pol_score
```

```python
#Calculating subjectivity score
def subjectivity_score(Total_words_after_clean,positiveScore, negativeScore):
    sub_score = (positiveScore + negativeScore) / ((Total_words_after_clean) + 0.000001)
    return sub_score
```

```python
#Calculating Average sentence length 
def average_sentence_length(text):
    sentence_list = sent_tokenize(text)
    tokens = tokenizer(text)
    totalWordCount = len(tokens)
    totalSentences = len(sentence_list)
    average_sent = 0
    if totalSentences != 0:
        average_sent = totalWordCount / totalSentences
    
    average_sent_length= average_sent
    
    return round(average_sent_length)
```

```python
#Calculating percentage of complex word 
def percentage_complex_word(text):
    tokens = tokenizer(text)
    complexWord = 0
    complex_word_percentage = 0
    
    for word in tokens:
        vowels=0
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            if(vowels > 2):
                complexWord += 1
    if len(tokens) != 0:
        complex_word_percentage = complexWord/len(tokens)
    
    return complex_word_percentage
```

```python
#Calculating Fog Index 
def fog_index(averageSentenceLength, percentageComplexWord):
    fogIndex = 0.4 * (averageSentenceLength + percentageComplexWord)
    return fogIndex
```

```python
#Average Number of Words Per Sentence
def avg_words_sentences(data):
  sent = sent_tokenize(data)
  len_sent = len(sent)
  w = tokenizer(data)
  len_w = len(w)
  avgtot = len_w / len_sent
  return avgtot
```

```python
#Counting complex words
def complex_word_count(text):
    tokens = tokenizer(text)
    complexWord = 0
    
    for word in tokens:
        vowels=0
        if word.endswith(('es','ed')):
            pass
        else:
            for w in word:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u'):
                    vowels += 1
            if(vowels > 2):
                complexWord += 1
    return complexWord
```

```python
#Counting syllable per word but first removed words ending with es or ed
def remove_esed_countvowels(text):
  str_tok = ' '.join([str(item) for item in text])
  syllRegex = re.compile(r'\b([A-z]+es|[A-z]+ed)\b',re.I)
  syll = syllRegex.findall(str_tok)
  tokenizer = RegexpTokenizer(r'\w+')
  tokens = tokenizer.tokenize(str_tok)
  filtered_words = list(filter(lambda token: token not in syll, tokens))
  new_str_toks = ' '.join([str(item) for item in filtered_words])
  string=new_str_toks
  vowels=0
  for i in string:
    if(i=='a' or i=='e' or i=='i' or i=='o' or i=='u' or i=='A' or i=='E' or i=='I' or i=='O' or i=='U'):
      vowels=vowels+1
  return vowels
```

```python
#Counting personal pronouns
import re
def personal_pronouns(text):
  tokens = tokenizer(text)
  str_toks = ' '.join([str(item) for item in tokens])
  pronounRegex = re.compile(r' I |we |We |My |my |Ours |ours |us ',re.I)
  pronouns = pronounRegex.findall(str_toks)
  prn = len(pronouns)
  return prn
```

```python
#Average word length
def avg_word_length(text):
  tokens = tokenizer(text)
  awl = sum(len(word) for word in tokens) / len(tokens)
  return awl
```

```python
#created a list for each value that is being computed to save the values 
positive_scores = []
negative_scores = []
polarity_scores = []
subjectivity_scores = []
average_sentence_lengths = []
percentage_complex_words = []
fog_indexs = []
average_number_of_words_per_sentences = []
complex_word_counts = []
total_word_counts = []
syllable_per_words = []
personal_pronouns_counts = []
average_word_lengths = []
```

```python
#iterating through each file and saving their computed values in their respective lists
for i in input['file_path']:
  with open(i, "r") as f:
    data = f.read()
    s = tokenizer(data)
    p = positive_score(s)
    positive_scores.append(p) 
    n = negative_word(s) 
    negative_scores.append(n)
    ps = polarity_score(p,n)
    polarity_scores.append(ps)
    Total_words_after_clean = len(s)
    sb = subjectivity_score(Total_words_after_clean,p,n)
    subjectivity_scores.append(sb)
    asl = average_sentence_length(data)
    average_sentence_lengths.append(asl)
    pcw = percentage_complex_word(data)
    percentage_complex_words.append(pcw)
    fi = fog_index(asl,pcw)
    fog_indexs.append(fi)
    avgtot = avg_words_sentences(data)
    average_number_of_words_per_sentences.append(avgtot)
    cwc = complex_word_count(data)
    complex_word_counts.append(cwc)
    twc = total_word_count(data)
    total_word_counts.append(twc)
    resed_cv = remove_esed_countvowels(s)
    syllable_per_words.append(resed_cv)
    pp = personal_pronouns(data)
    personal_pronouns_counts.append(pp)
    awl = avg_word_length(data)
    average_word_lengths.append(awl)
```

```python
#adding all the saved computed values into seperate columns
input['positive score'] = positive_scores
input['negative score'] = negative_scores
input['polarity score'] = polarity_scores
input['subjectivity score'] = subjectivity_scores
input['avg sentence length'] = average_sentence_lengths
input['percentage of complex words'] = percentage_complex_words
input['fog index'] = fog_indexs
input['avg number of words per sentence'] = average_number_of_words_per_sentences
input['complex word count'] = complex_word_counts
input['word count'] = total_word_counts
input['syllable per word'] = syllable_per_words
input['personal pronouns'] = personal_pronouns_counts
input['avg word length'] = average_word_lengths
input
```
